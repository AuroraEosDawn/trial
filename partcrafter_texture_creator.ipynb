```python
import os
import shutil

source_dir = "/content/drive/MyDrive/Colab/mira_input" # Adjust the path if your directory is elsewhere in Drive
destination_dir = "/content/mira_input"

# Check if the source directory exists in Google Drive
if os.path.exists(source_dir):
    # Remove the destination directory if it already exists to avoid issues
    if os.path.exists(destination_dir):
        print(f"Removing existing directory: {destination_dir}")
        shutil.rmtree(destination_dir)

    # Copy the directory
    print(f"Copying directory from: {source_dir} to {destination_dir}")
    shutil.copytree(source_dir, destination_dir)
    print("Copy complete.")
else:
    print(f"Error: Source directory not found at {source_dir}")
```

```python
import os
import glob
import subprocess

# Root directory where mira_input is extracted
input_root = '/content/mira_input'
inference_script = '/content/PartCrafter/scripts/inference_partcrafter.py'
# Removed the incorrect line: output_dir = "/content/drive/MyDrive/Colab/partcrafter_parts/", type=str, default="./results"
# Define the base output directory on Google Drive
base_output_dir = "/content/drive/MyDrive/Colab/partcrafter_parts/"

# Loop through all subfolders and .png images
for root, dirs, files in os.walk(input_root):
    for file in files:
        if file.endswith('.png') and not file.startswith('.'):
            img_path = os.path.join(root, file)
            # Create a tag based on subfolder and filename
            relative_path = os.path.relpath(img_path, input_root)
            tag = os.path.splitext(relative_path.replace("/", "_").replace(" ", "_"))[0]

            # Construct the specific output directory for this image based on the tag
            image_output_dir = os.path.join(base_output_dir, tag)

            # --- Add logic here to check if output already exists ---
            if os.path.exists(image_output_dir):
                print(f"✅ Skipping {img_path}: Output directory already exists at {image_output_dir}")
                continue # Skip to the next file if output exists
            # --- End of added logic ---

            print(f"🟢 Running PartCrafter on: {img_path} → Tag: {tag} → Output Dir: {image_output_dir}")

            # Ensure the output directory for this image exists before processing
            os.makedirs(image_output_dir, exist_ok=True)


            cmd = [
                'python3',
                inference_script,
                '--image_path', img_path,
                '--num_parts', '10', # adjust for complexity
                '--tag', tag,
                '--render',
                '--output_dir', image_output_dir # Pass the specific output directory
            ]

            try:
                # Set PYTHONPATH for the subprocess call
                env = os.environ.copy()
                env["PYTHONPATH"] = "/content/PartCrafter:" + env.get("PYTHONPATH", "")

                # Change to the PartCrafter directory before running the script
                # This is important for the script to find its internal modules
                subprocess.run(cmd, check=True, cwd="/content/PartCrafter", env=env)

            except subprocess.CalledProcessError as e:
                print(f"❌ Failed on {img_path}: {e}")
            except FileNotFoundError as e:
                # Handle case where inference_script might not be found
                 print(f"❌ Failed to find inference script: {e}")
```
